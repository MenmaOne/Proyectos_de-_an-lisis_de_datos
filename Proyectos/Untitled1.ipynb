```python
# Install and load necessary packages
install.packages("ggplot2")
install.packages("GGally")
```

```python
library(ggplot2)
library(GGally)

# Define function
explore_data <- function(df) {
  # Split the data frame by variable type
  num_vars <- sapply(df, is.numeric)
  quant_vars <- names(which(num_vars == TRUE))
  cat_vars <- names(which(num_vars == FALSE))
  
  # Descriptive statistics for quantitative variables
  cat("\n\nDescriptive Statistics for Quantitative Variables:\n")
  print(summary(df[, quant_vars]))
  
  # Bar plots for categorical variables
  cat("\n\nBar Plots for Categorical Variables:\n")
  for (var in cat_vars) {
    ggplot(df, aes_string(var)) +
      geom_bar() +
      ggtitle(paste("Bar Plot of", var)) +
      theme(plot.title = element_text(hjust = 0.5)) +
      labs(x = var, y = "Count")
  }
  
  # Pairwise scatter plots for quantitative variables
  cat("\n\nPairwise Scatter Plots for Quantitative Variables:\n")
  ggpairs(df[, quant_vars])
}

# Example usage
data(iris)
explore_data(iris)
```
