```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def describe_data(dataframe, vars_selected):
  # Descripción numérica de las variables del dataframe
  print(dataframe.describe())
  
  # Gráficos de barras para las variables categóricas
  cat_vars = dataframe.select_dtypes(include=['category']).columns
  for var in cat_vars:
    plt.figure()
    sns.countplot(x=var, data=dataframe)
    plt.title("Gráfico de barras para la variable '" + var + "'")
  
  # Gráficos de histogramas para las variables cuantitativas
  quant_vars = dataframe.select_dtypes(include=['float64', 'int64']).columns
  for var in quant_vars:
    plt.figure()
    sns.histplot(x=var, data=dataframe)
    plt.title("Histograma para la variable '" + var + "'")
  
  # Gráfico de dispersión por pares para las variables seleccionadas
  pairs_plot_vars = [var for var in vars_selected if var in quant_vars]
  sns.pairplot(dataframe[pairs_plot_vars])
```
