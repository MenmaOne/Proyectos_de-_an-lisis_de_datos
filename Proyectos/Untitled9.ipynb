```python
from graphviz import Digraph

# Crear el diagrama de actividades
phishing_diagram = Digraph(name="Phishing Process", format="png")
phishing_diagram.attr(rankdir="TB", size="8,10")

# Definir nodos de actividad
phishing_diagram.node("Start", "Inicio", shape="ellipse", style="filled", fillcolor="lightgreen")
phishing_diagram.node("TargetResearch", "Investigación del objetivo", shape="box", style="filled", fillcolor="lightblue")
phishing_diagram.node("CraftEmail", "Creación de correo falso", shape="box", style="filled", fillcolor="lightblue")
phishing_diagram.node("SendEmail", "Envío del correo de phishing", shape="box", style="filled", fillcolor="lightblue")
phishing_diagram.node("VictimAction", "Víctima interactúa con el enlace", shape="diamond", style="filled", fillcolor="yellow")
phishing_diagram.node("HarvestData", "Captura de datos personales", shape="box", style="filled", fillcolor="orange")
phishing_diagram.node("ExploitData", "Explotación de datos robados", shape="box", style="filled", fillcolor="red")
phishing_diagram.node("End", "Fin del ataque", shape="ellipse", style="filled", fillcolor="lightgray")

# Definir conexiones
phishing_diagram.edges([
    ("Start", "TargetResearch"),
    ("TargetResearch", "CraftEmail"),
    ("CraftEmail", "SendEmail"),
    ("SendEmail", "VictimAction"),
    ("VictimAction", "HarvestData", "Sí"),
    ("HarvestData", "ExploitData"),
    ("ExploitData", "End"),
    ("VictimAction", "End", "No")
])

# Guardar y renderizar el diagrama
phishing_diagram.render(filename="/mnt/data/phishing_diagram", cleanup=True)
"El diagrama se ha generado correctamente como archivo PNG."
```
