```python
# prompt: Cual  es el registro con la fecha mas lejana?

import pandas as pd


# Lee el archivo CSV en un DataFrame de pandas.
df = pd.read_csv("covid_19_data.csv")
  # Convierte la columna 'ObservationDate' a tipo datetime
df['ObservationDate'] = pd.to_datetime(df['ObservationDate'])

  # Encuentra el registro con la fecha más lejana
registro_mas_lejano = df.loc[df['ObservationDate'].idxmin()]

print("Registro con la fecha más lejana:")
print(registro_mas_lejano)
```
