```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data5 = pd.read_csv("pokemon.csv")
data5
```

```python
data5.info()
```

```python
df1 = data5.sort_values(by='Combat Power', ascending=False)
df1.head()
```

```python
top_5_1 = df1.head(5)
plt.figure(figsize=(10, 6))
plt.bar(top_5_1['Name'], top_5_1['Combat Power'], color='skyblue')
plt.xlabel('Nombre')
plt.ylabel('Poder de Combate')
plt.title('Top 5 Pokémon por Poder de Combate')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
