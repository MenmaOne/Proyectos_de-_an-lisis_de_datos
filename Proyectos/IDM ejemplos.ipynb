```python
#analisis exploratorio de datos: conjunto de datos iris

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

iris = load_iris()
```

```python
iris.data
```

```python
iris.target
```

```python
iris.feature_names
```

```python
df = pd.DataFrame(iris.data, columns= iris.feature_names)
df
```

```python
print(df.shape)
```

```python
df.info
```

```python
df.describe()
```

```python
df.describe().T
```

```python
#analisis univariable

fig, axes = plt.subplots(2,2, figsize = (10,8))

for i, ax in enumerate(axes.ravel()):
  ax.hist(df.iloc[:, i], bins = 15)
  ax.set_xlabel(df.columns[i])
  ax.set_ylabel('FRECUENCIA')
  ax.axvline(df.iloc[:, i].mean(), color = 'r', linestyle = 'dashed', linewidth = 1)
  ax.axvline(df.iloc[:, i].median(), color = 'g', linestyle = 'dashed', linewidth = 1)

plt.show()
```

```python
import seaborn as sns

df['target'] = iris.target
```

```python
sns.FacetGrid(df, hue='target', height=5).map(plt.scatter,'sepal length (cm)', 'sepal width (cm)').add_legend()
```

```python
sns.pairplot(df, hue = "target")

plt.show()
```

```python
sns.scatterplot(data = df, x = 'petal length (cm)', y = 'petal width (cm)', hue = 'target')
plt.show()
```

```python
dfx = df.drop(columns = ['target'])
sns.heatmap(dfx.corr(), annot=True, cmap = 'Blues')
plt.show()
```

```python
plt.scatter(df['sepal length (cm)'], df['sepal width (cm)'],  c = iris.target)
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.show()
```

```python
plt.scatter(df['petal length (cm)'], df['petal width (cm)'],  c = iris.target)
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')
plt.show()
```
