```python
from numpy import unique
from numpy import where
from sklearn.datasets import make_classification
from sklearn.cluster import AgglomerativeClustering
from matplotlib import pyplot
```

```python
X, _ = make_classification(n_samples=1000, n_features=2, n_informative=2, n_redundant=0, n_clusters_per_class=1, random_state=1)
```

```python
mdl = AgglomerativeClustering(n_clusters=2)
y_prima = mdl.fit_predict(X)
grupos = unique(y_prima)

for grupo in grupos:
  row_ix = where(y_prima == grupo)
  pyplot.scatter(X[row_ix, 0], X[row_ix, 1])

pyplot.show()
```

```python
import scipy.cluster.hierarchy as sch
#from scipy.cluster.hierarchy import dendogram, linkage

dendo = sch.dendrogram(sch.linkage(X, method = 'ward'))

```
